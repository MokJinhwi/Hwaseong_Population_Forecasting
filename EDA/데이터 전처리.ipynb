```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("최종 데이터.xlsx", index_col = '월별')
df.head
```

```python
df = df.drop(labels = '행정기관', axis = 1)
```

```python
df.head()
```

```python
data=[]
TCODE = [None]
data = df['총인구']
for i in range(len(data)-1):
    TCODE.append(np.log(data[i+1]) - np.log(data[i]))
TCODE = pd.DataFrame({'TCODE' : TCODE}, index = df.index)
```

```python
df = pd.concat([df, TCODE], axis = 1)
```

```python
df
```

```python
df.to_excel("last_data.xlsx")
```

```python
x = df.index
y1 = TCODE
```

```python
import matplotlib.pyplot as plt



fig = plt.figure(figsize = (12, 7))
ax = plt.subplot(111)

ax.plot(x, y1, color='b', label='mean_tem')


# 그래프 제목과 축 레이블 설정
ax.set_title('temperature')
ax.set_xlabel('Month')
ax.set_ylabel('Celsius')

# 그리드 표시
ax.grid(True)

# 범례 추가
ax.legend()

# 그래프 표시
plt.show()
```
